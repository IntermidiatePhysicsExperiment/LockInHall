```python
from packages import *
```

```python
data_path = '../DATA/B-2/'
file_path = lambda n: data_path + 'B-2_2kHz_1V_0.03s_6_10e-{}.csv'.format(n)
COLUMNS = [	'Phase difference [degree]',	'output [mV]']
def get_data(n):
  data = pd.read_csv(file_path(n))
  phase, output = data[COLUMNS[0]], data[COLUMNS[1]]
  phase = phase * np.pi / 180
  output = np.array(output)
  return phase, output

```

```python

phase, output = get_data(1)
x = phase.unique()
y = np.zeros((len(x), 2))
for idx, val in enumerate(x):
  y[idx, 0] = np.mean(output[phase == val])
  y[idx, 1] = np.std(output[phase == val])
plt.errorbar(x, y[:, 0], yerr=y[:, 1], fmt='o', ms=5, 
            color='green', label=f'S/N = $10^1$',
            capsize=5, capthick=1, )
func = lambda x, a, b: a * np.sin(x + b)
popt, pcov = scipy.optimize.curve_fit(func, phase, output)
plt.plot(phase, func(phase, *popt), 'g--')

phase, output = get_data(3)
x = phase.unique()
y = np.zeros((len(x), 2))
for idx, val in enumerate(x):
  y[idx, 0] = np.mean(output[phase == val])
  y[idx, 1] = np.std(output[phase == val])
plt.errorbar(x, y[:, 0], yerr=y[:, 1], fmt='o', ms=5, 
            color='red', label=f'S/N = $10^3$',
            capsize=5, capthick=1)
func = lambda x, a, b: a * np.sin(x + b)
popt, pcov = scipy.optimize.curve_fit(func, phase, output)
plt.plot(phase, func(phase, *popt), 'r--')

phase, output = get_data(5)
x = phase.unique()
y = np.zeros((len(x), 2))
for idx, val in enumerate(x):
  y[idx, 0] = np.mean(output[phase == val])
  y[idx, 1] = np.std(output[phase == val])
plt.errorbar(x, y[:, 0], yerr=y[:, 1], fmt='o', ms=5, 
            color='blue', label=f'S/N = $10^5$',
            capsize=5, capthick=1)
func = lambda x, a, b: a * np.sin(x + b)
popt, pcov = scipy.optimize.curve_fit(func, phase, output)
plt.plot(phase, func(phase, *popt), 'b--')

plt.legend()
plt.xlabel('Phase difference [rad]')
plt.ylabel('Lock-In Output [mV]')
plt.savefig('outputs/B-2/all.png')
```

```python
x = [1, 2, 5, 10, 20]
y = [1556, 1155, 839, 693, 468]
plt.plot(x, y, 'o-', color='black')
func = lambda x, a, b: a / np.power(10, x/20) + b
popt, pcov = scipy.optimize.curve_fit(func, x, y)
print(popt)
plt.plot(np.linspace(0.9, 20.3, 100), func(np.array(np.linspace(0.9, 20.3, 100)), *popt), 'r--')
R2 = 1 - np.sum((y - func(np.array(x), *popt))**2) / np.sum((y - np.mean(y))**2)
print(R2)
plt.text(17, 1500, f'$R^2$ = {R2:.3f}')
plt.xlabel('Gain [dB]')
plt.ylabel('Frequency [Hz]')
plt.title('Frequency vs Gain')
```

```python
x = [0.03, 0.1, 0.3, 0.03, 0.1, 0.3]
y = [34.3, 12.5, 5.08, 21.5, 9.19, 1.90]
plt.scatter(x, y, color='black')
func = lambda x, a: 1 / x
popt, pcov = scipy.optimize.curve_fit(func, x, y)
print(popt)
R2 = 1 - np.sum((y - func(np.array(x), *popt))**2) / np.sum((y - np.mean(y))**2)
print(R2)
plt.plot(np.linspace(0.02, 0.32, 100), func(np.array(np.linspace(0.02, 0.32, 100)), *popt), 'r--')
plt.xlabel('Time Constant [s]')
plt.ylabel('Frequency [Hz]')
plt.title('Frequency vs Time Constant')
plt.text(0.27, 40, f'$R^2$ = {R2:.3f}')
```
